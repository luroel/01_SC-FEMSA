```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lectura de datos del archivo Storecheck_Enfriador_Respeto_Distribuidores_data.xlsm, hoja Comportamiento Enfriadores Geo
storecheck_df = pd.read_excel('/Users/luigi/Documents/DataScience_REMOTEdata_Github/01_SC-FEMSA/data/Storecheck_Enfriador_Respeto_Distribuidores_data.xlsm', sheet_name='Comportamiento Enfriadores Geo')
storecheck_df.head(7)
```

```python
# Total de elementos Mes 1 por Categoría Levantamientos
levantamientos_mes1 = storecheck_df[(storecheck_df['Categoría'] == 'Levantamientos')]['Mes 1'].sum()
# Total de elementos Mes 2 por Categoría Levantamientos
levantamientos_mes2 = storecheck_df[(storecheck_df['Categoría'] == 'Levantamientos')]['Mes 2'].sum()
# Total elementos vs MA Levantamientos
levantamientos_vsma = storecheck_df[(storecheck_df['Categoría'] == 'Levantamientos')]['Vs MA'].sum()
print(f"Total Levantamientos Mes 1: {levantamientos_mes1}")
print(f"Total Levantamientos Mes 2: {levantamientos_mes2}")
print(f"Total Vs MA Levantamientos: {levantamientos_vsma}")
```

```python
# Total de elementos Mes 1 por Categoría Refrigeradores
refrigeradores_mes1 = storecheck_df[(storecheck_df['Categoría'] == 'Refrigeradores')]['Mes 1'].sum()
# Total de elementos Mes 2 por Categoría Refrigeradores
refrigeradores_mes2 = storecheck_df[(storecheck_df['Categoría'] == 'Refrigeradores')]['Mes 2'].sum()
# Total elementos vs MA Refrigeradores
refrigeradores_vsma = storecheck_df[(storecheck_df['Categoría'] == 'Refrigeradores')]['Vs MA'].sum()
print(f"Total Refrigeradores Mes 1: {refrigeradores_mes1}")
print(f"Total Refrigeradores Mes 2: {refrigeradores_mes2}")
print(f"Total Vs MA Refrigeradores: {refrigeradores_vsma}")
```
