```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lectura de datos del archivo Storecheck_Enfriador_Respeto_Distribuidores_data.xlsm, hoja Comportamiento Enfriadores Geo
df_storecheck = pd.read_excel('/Users/luigi/Documents/DataScience_REMOTEdata_Github/01_SC-FEMSA/data/Storecheck_Enfriador_Respeto_Distribuidores_data.xlsm', sheet_name='Comportamiento Enfriadores Geo')
df_storecheck.head(7)
```

```python
# Total de elementos Mes 1 por Categoría Levantamientos
levantamientos_mes1 = df_storecheck[(df_storecheck['Categoría'] == 'Levantamientos')]['Mes 1'].sum()
# Total de elementos Mes 2 por Categoría Levantamientos
levantamientos_mes2 = df_storecheck[(df_storecheck['Categoría'] == 'Levantamientos')]['Mes 2'].sum()
# Total elementos vs MA Levantamientos
levantamientos_vsma = df_storecheck[(df_storecheck['Categoría'] == 'Levantamientos')]['Vs MA'].sum()
print(f"Total Levantamientos Mes 1: {levantamientos_mes1}")
print(f"Total Levantamientos Mes 2: {levantamientos_mes2}")
print(f"Total Vs MA Levantamientos: {levantamientos_vsma}")
```

```python
# Total de elementos Mes 1 por Categoría Refrigeradores
refrigeradores_mes1 = df_storecheck[(df_storecheck['Categoría'] == 'Refrigeradores')]['Mes 1'].sum()
# Total de elementos Mes 2 por Categoría Refrigeradores
refrigeradores_mes2 = df_storecheck[(df_storecheck['Categoría'] == 'Refrigeradores')]['Mes 2'].sum()
# Total elementos vs MA Refrigeradores
refrigeradores_vsma = df_storecheck[(df_storecheck['Categoría'] == 'Refrigeradores')]['Vs MA'].sum()
print(f"Total Refrigeradores Mes 1: {refrigeradores_mes1}")
print(f"Total Refrigeradores Mes 2: {refrigeradores_mes2}")
print(f"Total Vs MA Refrigeradores: {refrigeradores_vsma}")
```

```python
# Total de elementos Mes 1 por Categoría Lleno
lleno_mes1 = df_storecheck[(df_storecheck['Categoría'] == 'Lleno')]['Mes 1'].sum()
# Total de elementos Mes 2 por Categoría Lleno
lleno_mes2 = df_storecheck[(df_storecheck['Categoría'] == 'Lleno')]['Mes 2'].sum()
# Total elementos vs MA Lleno
lleno_vsma = df_storecheck[(df_storecheck['Categoría'] == 'Lleno')]['Vs MA'].sum()
print(f"Total Llenos Mes 1: {lleno_mes1}")
print(f"Total Llenos Mes 2: {lleno_mes2}")
print(f"Total Vs MA Llenos: {lleno_vsma}")
```

```python
# Total de elementos Mes 1 por Categoría Inavasiones a Refrigerador
invasionesrefrigerador_mes1 = df_storecheck[(df_storecheck['Categoría'] == 'Invasiones a refrigerador')]['Mes 1'].sum()
# Total de elementos Mes 2 por Categoría Inavasiones a Refrigerador
invasionesrefrigerador_mes2 = df_storecheck[(df_storecheck['Categoría'] == 'Invasiones a refrigerador')]['Mes 2'].sum()
# Total elementos vs MA Inavasiones a Refrigerador
invasionesrefrigerador_vsma = df_storecheck[(df_storecheck['Categoría'] == 'Invasiones a refrigerador')]['Vs MA'].sum()
print(f"Total Invasiones a Refrigerador Mes 1: {invasionesrefrigerador_mes1}")
print(f"Total Invasiones a Refrigerador Mes 2: {invasionesrefrigerador_mes2}")
print(f"Total Vs MA Invasiones a Refrigerador: {invasionesrefrigerador_vsma}")
```

```python
# Total de elementos Mes 1 por Categoría Porcentaje refris invadidos
Porcentaje_Invadidos_mes1 = df_storecheck[(df_storecheck['Categoría'] == 'Porcentaje refris invadidos (%)')]['Mes 1'].sum()
# Total de elementos Mes 2 por Categoría Porcentaje refris invadidos
Porcentaje_Invadidos_mes2 = df_storecheck[(df_storecheck['Categoría'] == 'Porcentaje refris invadidos (%)')]['Mes 2'].sum()
# Total elementos vs MA Porcentaje refris invadidos
Porcentaje_Invadidos_vsma = df_storecheck[(df_storecheck['Categoría'] == 'Porcentaje refris invadidos (%)')]['Vs MA'].sum()
print(f"Total Porcentaje refris invadidos Mes 1: {Porcentaje_Invadidos_mes1}%")
print(f"Total Porcentaje refris invadidos Mes 2: {Porcentaje_Invadidos_mes2}%")
print(f"Total Vs MA Porcentaje refris Invadidos: {Porcentaje_Invadidos_vsma}%")
```

# Indicadores Principales Comportamiento Enfriadores por Geografía
### Aumento significativo de las invasiones a refrigeradores de un mes a otro en todas las zonas.

```python
# barplot - Indicadores Principales del PDF EJERCICIO INVASION (zona 1-7)
mes = ['Mes 1', 'Mes 2']
Indicadores_Mensuales_Refris = [refrigeradores_mes1, refrigeradores_mes2]
Indicadores_Mensuales_Lleno = [lleno_mes1, lleno_mes2]
Indicadores_Mensuales_Invasiones = [invasionesrefrigerador_mes1, invasionesrefrigerador_mes2]
bar_width = 0.25
x = np.arange(len(mes))

# Grouped Bar Plot
plt.bar(x - bar_width, Indicadores_Mensuales_Refris, bar_width, label='Refrigeradores', color='skyblue')
plt.bar(x, Indicadores_Mensuales_Lleno, bar_width, label='Refrigeradores Llenos', color='lightgreen')
plt.bar(x + bar_width, Indicadores_Mensuales_Invasiones, bar_width, label='Invasiones', color='lightcoral')


# Adding labels and title
plt.ylabel('Cantidad')
plt.title('Comportamiento Enfriadores por Geografía (Zona 1-7)')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)
plt.tight_layout()
plt.xticks(x, mes)
plt.legend(title='Indicadores')
plt.show()
```
